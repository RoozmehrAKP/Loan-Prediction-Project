### We used a method called smote and split our data for further process

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
import warnings
from imblearn.under_sampling import RandomUnderSampler
warnings.filterwarnings("ignore", category=FutureWarning)

# Separating features (X) and target (y)
X = df.drop('Loan_Status', axis=1)
y = df['Loan_Status']

# Checking class distribution before resampling
print("Before balancing:")
print(Counter(y))

# Splitting into train-test sets
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Applying SMOTE to balance the training data
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(Xtrain, ytrain)
# Applying RandomOverSampler to balance the training data
# oversampler = RandomOverSampler(sampling_strategy={0: 414, 1: 414}, random_state=42)
# X_train_oversampled, y_train_oversampled = oversampler.fit_resample(Xtrain, ytrain)


# Applying RandomUnderSampler to balance the training data
# undersampler = RandomUnderSampler(random_state=42)
# X_train_undersampled, y_train_undersampled = undersampler.fit_resample(Xtrain, ytrain)
# X_test_undersampled, y_test_undersampled = undersampler.fit_resample(Xtest, ytest)

# Checking class distribution after SMOTE
print("After balancing:")
# print(Counter(y_train_smote))
print(Counter(y_train_smote))
```

```python
# from sklearn.neighbors import KNeighborsClassifier
# #Setup arrays to store training and test accuracies
# neighbors = np.arange(1,30)
# train_accuracy = np.empty(len(neighbors))
# test_accuracy = np.empty(len(neighbors))

# for i,k in enumerate(neighbors):
#     #Setup a knn classifier with k neighbors
#     knn = KNeighborsClassifier(n_neighbors=k)
    
#     #Fit the model
#     knn.fit(X_train_oversampled, y_train_oversampled)
    
#     #Compute accuracy on the training set
#     train_accuracy[i] = knn.score(X_train_oversampled, y_train_oversampled)
    
#     #Compute accuracy on the test set
#     test_accuracy[i] = knn.score(Xtest, ytest) 
```

```python
from sklearn.neighbors import KNeighborsClassifier
#Setup arrays to store training and test accuracies
neighbors = np.arange(1,30)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i,k in enumerate(neighbors):
    #Setup a knn classifier with k neighbors
    knn = KNeighborsClassifier(n_neighbors=k)
    
    #Fit the model
    knn.fit(X_train_smote, y_train_smote)
    
    #Compute accuracy on the training set
    train_accuracy[i] = knn.score(X_train_smote, y_train_smote)
    
    #Compute accuracy on the test set
    test_accuracy[i] = knn.score(Xtest, ytest) 
```

```python
#Generate plot
plt.title('k-NN Varying number of neighbors')
plt.plot(neighbors, test_accuracy, label='Testing Accuracy')
plt.plot(neighbors, train_accuracy, label='Training accuracy')
plt.legend()
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.show()
```
